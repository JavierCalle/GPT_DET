## preparar dataframe

```python
import pandas as pd
```

```python
textos_gpt = pd.read_csv('textos_gpt.csv')
```

```python
textos_gpt
```

```python
df1 = pd.read_csv('artlimpio.csv')
df2 = pd.read_csv('wiki a.csv')
```

```python
textos_humanos = pd.concat([df1, df2], ignore_index=True)
```

```python
textos_humanos
```

```python
textos_humanos['esia'] = 0
```

```python
textos_gpt['esia'] = 1
```

```python
textos_gpt = textos_gpt.rename(columns={'response': 'text'})
```

```python
result = pd.concat([textos_gpt, textos_humanos], ignore_index=True)
```

```python
result
```

```python
result.to_csv('entrenamiento_nr.csv', index=False)
```

```python
textos_humanos.to_csv('textoshumanos.csv', index=False)
```

```python
df_randomized = result.sample(frac=1).reset_index(drop=True)
```

```python
df_randomized
```

```python
df_randomized.to_csv('entrenamiento_r.csv', index=False)
```
