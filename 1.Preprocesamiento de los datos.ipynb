```python
import numpy as np
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('all')
```

```python
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # Convertir a minúsculas
    text = text.lower()
    
    # Eliminar signos de puntuación
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenización
    tokens = word_tokenize(text)
    
    # Eliminar stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    
    # Lematización
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)
```

```python
entrenamiento['text'] = entrenamiento['text'].apply(preprocess)
```

Procesamiento de texto uno a uno

```python
entrenamiento['text'] = entrenamiento['text'].str.lower()
```

```python
entrenamiento['text'] = entrenamiento['text'].apply(lambda x: re.sub(r'[^\w\s]|\\', '', x))
```

```python
entrenamiento['text'] = entrenamiento['text'].str.replace('\n', ' ')
```

```python
entrenamiento['text'] = entrenamiento['text'].apply(nltk.word_tokenize)
```

EL objetivo es detectar si un texto ha sido generado por una IA, no eliminare las stopwords en la fase inicial. Estas palabras pueden ofrecer pistas valiosas sobre la naturaleza del texto. Sin embargo, comparar los resultados para determinar qué enfoque funciona mejor.

Para el propósito específico de detectar textos generados por IA, podría ser más prudente comenzar sin lematizar el texto, ya que querrás capturar todas las posibles peculiaridades y características del texto original.

```python
from nltk.stem import WordNetLemmatizer

# Inicializa el lematizador
lemmatizer = WordNetLemmatizer()

# Función para lematizar un texto
def lematize_text_nltk(text):
    words = nltk.word_tokenize(text)
    return ' '.join([lemmatizer.lemmatize(word) for word in words])

# Aplica la función a la columna 'text'
entrenamiento['text'] = entrenamiento['text'].apply(lematize_text_nltk)
```

```python
entrenamiento['text'] = entrenamiento['text'].fillna('')
```

```python
entrenamiento['text'] = entrenamiento['text'].astype(str)
```
